## How to label customers whose aggregated `amount_spent` is in a given quantile range?


We break down the main problem into several sub problems:

#### Sub Problem 1: How to aggregate the  `amount_spent` for unique customers?

#### Sub Problem 2: How to select customers whose aggregated `amount_spent` is in a given quantile range?

#### Sub Problem 3: How to label selected customers as "VIP" or "Preferred"?

*Note: If you want to break down the main problem in a different way, please feel free to revise the sub problems above.*

Now in the workspace below, tackle each of the sub problems using the iterative problem solving workflow. Insert cells as necessary to write your codes and explain your steps.

```python
# your code here
# 1.
orders_agg = orders[['CustomerID','amount_spent']]
orders_agg = orders_agg.groupby('CustomerID')['amount_spent'].agg(total_spent=sum).reset_index()
print(orders_agg)
# 2.
display(orders_agg[orders_agg.total_spent > orders_agg.total_spent.quantile(.95)])
# 3.
def Client_type(row):
    if row.total_spent > orders_agg.total_spent.quantile(.95):
        type_c = "VIP"
    elif row.total_spent > orders_agg.total_spent.quantile(.75):
        type_c = "Preferred"
    else:
        type_c = "-"
    return type_c

orders_agg['Client_type'] = orders_agg.apply(Client_type, axis=1)
orders_agg
```

Now we'll leave it to you to solve Q2 & Q3, which you can leverage from your solution for Q1:

## Q2: How to identify which country has the most VIP Customers?

```python
sum(orders_agg.Client_type == 'VIP')
```

```python
# your code here
orders_agg_VIP = orders_agg[orders_agg.Client_type == 'VIP']
orders_country = orders[['CustomerID','Country']]
# display(orders_agg_VIP)
orders_mix = orders_agg_VIP.merge(orders_country, how = 'inner', left_on = "CustomerID", right_on= "CustomerID")
orders_mix = orders_mix.groupby(['Country','Client_type']).size().reset_index(name='counts')
orders_mix = orders_mix[['Country','counts']]
# display(orders_mix)
orders_mix[orders_mix.counts == orders_mix.counts.max()]
```

## Q3: How to identify which country has the most VIP+Preferred Customers combined?

```python
# your code here
orders_agg_VP = orders_agg[orders_agg.Client_type.isin(['VIP','Preferred'])]
orders_mix = orders_agg_VP.merge(orders_country, how = 'inner', left_on = "CustomerID", right_on= "CustomerID")
orders_mix = orders_mix.groupby(['Country','Client_type']).size().reset_index(name='counts')
orders_mix = orders_mix[['Country','counts']]
orders_mix[orders_mix.counts == orders_mix.counts.max()]
```
